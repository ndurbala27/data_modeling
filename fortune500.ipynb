```python
group_by_year = df.loc[:, ['year', 'revenue', 'profit']].groupby('year')
```

```python
avgs = group_by_year.mean()
```

```python
x = avgs.index
y1 = avgs.profit
```

```python
def plot(x, y, ax, title, y_label):
    ax.plot(x, y)
    ax.set_title(title)
    ax.set_ylabel(y_label)
    ax.margins(x=0, y=0)
```

```python
fig, ax = plt.subplots()
```

```python
plot(x, y1, ax, 'Increase in mean Fortune 500 company profits from 1955 to 2005', 'Profit (in millions)')
```

```python
plt.show()
```

```python
y2 = avgs.revenue
fig, ax = plt.subplots()
plot(x, y2, ax, 'Increase in mean Fortune 500 company revenue from 1955 to 2005', 'Revenue (in millions)')
plt.show()
```

```python
def plot_with_std(x, y, stds, ax, title, y_label):
    ax.fill_between(x, y - stds, y + stds, alpha=0.2)
    plot(x, y, ax, title, y_label)

fig, ax = plt.subplots()
title = 'Incerease in mean and std Fortune 500 company %s from 1955 to 2005'
stds1 = group_by_year.std().profit.values
stds2 = group_by_year.std().revenue.values
plot_with_std(x, y1, stds1, ax, title % 'profits', 'Profit (in millions)')
fig.set_size_inches(14,4)
fig.tight_layout()
plt.show()
```

```python
import ipywidgets as widgets
from IPython.display import display

def update_plot(year_range):
    start_year, end_year = year_range
    mask = (x >= start_year) & (x <= end_year)

    fig, ax = plt.subplots()
    title = 'Incerease in mean Fortune 500 company from %d to %d'
    plot(x[mask], y1[mask], ax, title % (start_year, end_year), 'Profit (in millions)')
    plt.show()
```

```python
year_range_slider = widgets.IntRangeSlider(
    value=[1955, 2005],
    min=1955,
    max=2005,
    step=1,
    description='Year Range',
    continuous_update=False
)
widgets.interactive(update_plot, year_range=year_range_slider)

```
